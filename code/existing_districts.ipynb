```python
def calc_RD(i, k, Graph_object): 
    node = Graph_object.nodes[i]
    list_of_neighbors= list(Graph_object.adj[i])
    num_neighbors = len(list_of_neighbors)
    if k == 2: 
        sum1=0
        for j in range(0,num_neighbors): 
            n_node_i= list_of_neighbors[j] #want to compare for W/ nW
            n_node= Graph_object.nodes[n_node_i]
            percent_div= n_node["n_white"]/n_node["tot18"]
            sum1 = sum1 + percent_div
        return(node["n_white"]/node["tot18"]- (sum1/num_neighbors)) # a score of zero means fits perfectly, a negative score means more diversity in surrounding area, a post score means more diversity in center node
    
    
    
```

```python
def all_nodes(graph_object): 
    rd_list = []
    for i in range (0,331): 
        rd_list.append(calc_RD(i, 2, graph_object))
    return rd_list
```

```python
def all_nodes_info(graph_object): 
    rd_list = []
    for i in range (0,331): 
        rd_list.append((calc_RD(i, 2, graph_object),graph_object.nodes[i]['index']))
    return rd_list
```
